```python
import PyPDF2
import pdfplumber
import os
import re
import random
import spacy
```

```python
pdf_directory = 'Testing_pdf_data'
txt_directory = 'txt_pdf_files'
if not os.path.exists(txt_directory):
    os.makedirs(txt_directory)
```

#### Converting .pdf to .txt

```python
if not os.path.exists(txt_directory):
    os.makedirs(txt_directory)

def convert_pdf_to_txt(pdf_file):
    try:
        print(pdf_directory+ '/' + pdf_file)
        with pdfplumber.open(pdf_directory+ '/' + pdf_file) as pdf:
            text = ''
            for page in pdf.pages:
                text += page.extract_text()
            return text
    except Exception as e:
        print(f"Error converting {pdf_file} to TXT: {str(e)}")
        return None

str1 = ['\n', '‘', '’' ,' ','!','"','&',"'",'(',')',',','-', '.', ':', ';', '?', '[', ']', '_', 'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']


```

```python
# Loop through all PDF files in the PDF directory
for pdf_file in os.listdir(pdf_directory):
    if pdf_file.endswith('.pdf'):
        # Convert the PDF to TXT
        txt_content = convert_pdf_to_txt(pdf_file)
        txt_content = txt_content.lower()
        txt_content = re.sub(r'\s+', ' ', txt_content)
        cleaned_text = ''.join(char for char in txt_content if char in str1)
    
        # Write train data to filename_train.txt
        txt_filename = pdf_file.replace('.pdf', '.txt')
        with open(f'{txt_directory}/{txt_filename}_org.txt', 'w', encoding='utf-8') as train_file:
            train_file.writelines(cleaned_text)

print("Conversion completed. TXT files saved in the specified directory.")
```

#### Combining all books data to one txt file

```python
def split_data(filename):

    nlp = spacy.load("en_core_web_sm",disable=["parser", "ner"])
    if not nlp.has_pipe("sentencizer"):
        nlp.add_pipe("sentencizer")
    nlp.max_length = 1500000

    # Read the content of the input file
    with open(filename, 'r', encoding='utf-8') as file:
        content = file.read()
    
    doc = nlp(content)
    sentences = [sent.text for sent in doc.sents]
    name = filename.split('/')[-1].split('.')[0]
    
    print(name)
   
    # Calculate the number of sentences for test data (20%)
    num_sent = int(len(sentences) * 0.2)
    test_data = random.sample(sentences, num_sent)
    
    train_data = [sent for sent in sentences if sent not in test_data]
    train_data = ' '.join(train_data)
    test_data = ' '.join(test_data)

    # Write train data to filename_train.txt
    with open(f'txt_pdf_files/All_new_train1.txt', 'a', encoding='utf-8') as train_file:
        train_file.write(train_data + '\n')

    # Write test data to filename_test.txt
    with open(f'txt_pdf_files/All_new_test1.txt', 'a', encoding='utf-8') as test_file:
        test_file.write(test_data + '\n')
```

```python
lst = os.listdir(txt_directory)
lst.remove('All_new_train.txt')
lst.remove('All_new_test.txt')
lst.remove('All_new_train1.txt')
lst.remove('All_new_test1.txt')
```
